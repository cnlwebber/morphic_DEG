## Read in the counts and meta data

```r
######## Read in raw count file ########
raw_cnts = read.csv("./final_project/GSE288288_study1_genesRawCounts.csv", header=TRUE, sep="\t", row.names=1)
head(raw_cnts)
dim(raw_cnts)
```

```r
######## Read in metadata file ########
meta_data = read.csv("./final_project/metadata_study1.csv", header=TRUE, stringsAsFactors=FALSE)
head(meta_data)
dim(meta_data)
```

```r
colnames(meta_data) <- gsub(" ", "_", colnames(meta_data))
colnames(meta_data)
```

```r
rownames(meta_data) = meta_data$Sample
# trim the ends 
colnames(raw_cnts) <- gsub("_L\\d{3}$", "", colnames(raw_cnts))
# replace dot (.) to hyphen (-)
colnames(raw_cnts) <- gsub("\\.", "-", colnames(raw_cnts))
table(rownames(meta_data) %in% colnames(raw_cnts))
raw_cnts = raw_cnts[,rownames(meta_data)]
dim(raw_cnts)
dim(meta_data)
head(raw_cnts)
head(meta_data)
```
